```python
ss=StandardScaler()
x=ss.fit_transform(x)
x
```

```python
#hierarchical clustering
```

```python
plt.figure(figsize=(10,7))
plt.title("dendogram")
dend=shc.dendrogram(shc.linkage(x,method="complete"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg=AgglomerativeClustering(n_clusters=2,metric="euclidean",linkage="complete")
y=agg.fit_predict(x)
y=pd.DataFrame(y)
y.value_counts()
```

```python
# Scatter plot for hierarchical clustering
plt.figure(figsize=(8, 6))
plt.scatter(x[y[0]==0][:, 0], x[y[0]==0][:, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[y[0]==1][:, 0], x[y[0]==1][:, 1], s=50, c='blue', label='Cluster 2')
plt.title('Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
'''The hierarchical clustering algorithm divided the data into two clusters (red and blue).
Points in the red cluster are close to each other, indicating a higher similarity'''
```

```python
 #kmeans clustering
```

```python
# Find the optimal number of clusters using the elbow method
wcss = []#inertia
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(x, kmeans.labels_))
```

```python
# Plot the elbow method graph
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
```

```python
from sklearn.cluster import KMeans
KMeans = KMeans(n_clusters = 4, n_init=30)
KMeans.fit(x)
Y = KMeans.predict(x)
Y = pd.DataFrame(Y)
Y[0].value_counts()
```

```python
# Scatter plot for k-means clustering
plt.figure(figsize=(8, 6))
plt.scatter(x[Y[0]==0][:, 0], x[Y[0]==0][:, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[Y[0]==1][:, 0], x[Y[0]==1][:, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(x[Y[0]==2][:, 0], x[Y[0]==2][:, 1], s=50, c='green', label='Cluster 3')
plt.scatter(x[Y[0]==3][:, 0], x[Y[0]==3][:, 1], s=50, c='purple', label='Cluster 4')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
'''The k-means algorithm with k=4 divided the data into four clusters (red, blue, green, and purple).
Each cluster represents a distinct group of points with similar characteristics.'''
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.70,min_samples=3)
dbscan.fit(x)
cl=pd.DataFrame(dbscan.labels_,columns=["cluster"])
cl.value_counts()
clustered=pd.concat([d1,cl],axis=1)
noise_data=clustered[clustered["cluster"]==-1]
final_data=clustered[clustered["cluster"]==0]
```

```python
# Scatter plot for DBSCAN
plt.figure(figsize=(8, 6))
plt.scatter(x[cl['cluster']==0][:, 0], x[cl['cluster']==0][:, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[cl['cluster']==1][:, 0], x[cl['cluster']==1][:, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(x[cl['cluster']==-1][:, 0], x[cl['cluster']==-1][:, 1], s=50, c='grey', label='Noise')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
'''DBSCAN identified two clusters (red and blue) and labeled some points as noise (grey).
Points in the red and blue clusters are well-defined and separated.
The grey points represent noise or outliers that do not belong to any specific cluster.'''
```
